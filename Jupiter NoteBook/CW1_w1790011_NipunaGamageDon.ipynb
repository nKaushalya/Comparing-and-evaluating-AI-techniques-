### Artificial Neural Network (ANN)

##### ANN uses brain processing to create algorithms that can be used to model complex patterns and predict problems. An artificial neural network is made up of three or more interconnected layers. The first layer is made up of input neurons. Those neurons send data to the deeper layers, which send the final output data to the
final output layer. The inner layers are all hidden and formed by units that change the information received from layer to layer adaptively through a series of transformations. Each layer serves as both an
input and output layer, allowing the ANN to comprehend more complex objects. These inner layers are referred to collectively as the neural layer. The neural layer units attempt to learn about the information gathered by weighing it according to the ANN's internal system. These guidelines enable units to generate a transformed result, which is then provided as an output to the next layer.

Advantages
1. Having a distributed memory: In order for an ANN to learn, examples must be determined and
taught to the network based on the desired output by displaying these examples to the network.
The network&#39;s success is directly proportional to the number of instances selected, and if the
event cannot be shown to the network in all of its aspects, the network may produce incorrect
output.
2. Gradual deterioration: A network slows down and degrades over time. The network problem does
not manifest itself immediately.
3. Machine learning capability: Artificial neural networks learn events and make decisions by
commenting on similar events.

Disadvantages
1. Determination of proper network structure: There are no hard and fast rules for determining the
structure of artificial neural networks. Appropriate network structure is attained through trial and
error.
2. Difficulty in communicating the problem to the network: ANNs can work with numerical data.
Before introducing problems to ANN, they must be converted into numerical values. The display
mechanism chosen here will have a direct impact on network performance. This is determined by
the user's ability.

### Random Forest Classifier

An ensemble approach based on decision trees is called Random Forest. By integrating various overfit
evaluators (such as decision trees) to create an ensemble learning method, RF lessens the degree of
overfitting. The relevant classification decision result can be obtained for each decision tree. The
category of the sample to be tested is decided using the voting outcomes of each decision tree in the
forest in accordance with the maxim of the minority obeying the majority, and the category with higher
votes across all decision trees was found to be the outcome. The way to identify heart illness using
Python in this project uses the random forest algorithm, which is used for developing heart disease

### Gradient Boosting

Granting a boost, the basic concept is to first initialize the model and establish the loss function,
calculate the pseudo residual under the original model, create a learner to explain the pseudo residual,
reduce the pseudo residual in the gradient direction, multiply the learner by the weight coefficient and
linearly combine with the original model to create a new model, and repeat the process as necessary.
For the model to move in the direction of the fastest residual reduction and produce a succession of
weak classifiers, each of which is a binary tree, it is essential that each new iteration lower the residual
of the preceding iteration. The combination of these weak classifiers would result in a model that could
reduce the loss function.
The gradient boosting learning approach originates from the gradient boosting decision tree.
Generalization-wise, GB performs well as an ensemble classifier. Additionally, GB offers a regularization
term to control the complexity of the model, preventing overfitting. GB has outperformed the
competition in a number of machine learning fields.

Advantages

1. Frequently offers unrivalled forecasting accuracy.
2. The function fit has a lot of versatility because it can be optimized for a variety of loss functions
and offers a number of hyperparameter adjustment possibilities.

3. No pre-processing of the data is necessary; category and numerical values frequently perform
well when presented as-is.
4. Deals with missing data; imputation is not necessary.

Disadvantages

1. GBMs will continue to improve in order to reduce all errors. This can lead to overfitting by
exaggerating outliers. To neutralize, cross-validation must be used.
2. Because of the approach&#39;s high flexibility, many parameters interact and heavily influence its
behavior (number of iterations, tree depth, regularization parameters, etc.). During tuning, this
necessitates a large grid search.
3. Less interpretable, though this is easily remedied with a variety of tools

# Modelling

```python
#Train / Test Split with preserving outcome ratio using Stratify

from sklearn.model_selection import train_test_split

train,test= train_test_split(Heart_Data,test_size=0.25,random_state=0,stratify=Heart_Data['target'])# stratify the outcome

train_X=train[train.columns[:13]]
test_X=test[test.columns[:13]]
train_Y=train['target']
test_Y=test['target']
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 80, num = 10)]
# Number of features to consider at every split
#max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [2,4]
# Minimum number of samples required to split a node
min_samples_split = [2, 5]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2]
# Method of selecting samples for training each tree
bootstrap = [True, False]
```

```python
# Create the param grid
param_grid = {'n_estimators': n_estimators,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(param_grid)
```

```python
#Define the Classifier
from sklearn.ensemble import RandomForestClassifier 
rf_Model_Hp = RandomForestClassifier()
```

```python
#Perform the RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV
rf_Random_Grid = RandomizedSearchCV(estimator = rf_Model_Hp, param_distributions = param_grid, cv = 10, verbose=2, n_jobs = 4)
```

```python
#Training Outcome
rf_Random_Grid.fit(train_X, train_Y)
```

```python
#Finalized Best params
rf_Random_Grid.best_params_
```

```python
#Check the Accuracy
print (f'Train Accuracy - : {rf_Random_Grid.score(train_X,train_Y):.3f}')
print (f'Test Accuracy - : {rf_Random_Grid.score(test_X,test_Y):.3f}')
```

```python
# Random Forest Model

from sklearn.ensemble import RandomForestClassifier 
from sklearn import metrics
model_1= RandomForestClassifier(n_estimators = 41,
 min_samples_split = 2,
 min_samples_leaf = 2,
 max_depth = 4,
 bootstrap = True)
model_1.fit(train_X,train_Y)
p = model_1.predict(test_X)

# Results
print('The accuracy Score is:\n',metrics.accuracy_score(p,test_Y))
```

```python
# Accuracy Results
print('The accuracy Score is:\n',metrics.accuracy_score(p,test_Y))
print('\n \n The confusion matrix: \n', metrics.confusion_matrix(p, test_Y))
print('\n\n The metrics classification report:\n ', metrics.classification_report(p, test_Y))
```

```python
# Define ROC Curve

def plot_roc_curve(fpr, tpr):
    plt.plot(fpr, tpr, color='yellow', label='ROC')
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend()
    plt.show()


fpr, tpr, thresholds = metrics.roc_curve(test_Y, prob)
plot_roc_curve(fpr, tpr)
```
